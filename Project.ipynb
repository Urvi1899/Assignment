```python
import datetime

class Person:
    def __init__(self, name, dob):
        self.name = name
        self.dob = dob
    
    def get_approx_age(self):
        """
        Returns the approximate age of the person based on their date of birth.
        
        Assumptions:
        - The person's date of birth is provided as a string in the format 'YYYY-MM-DD'.
        - The person's age is calculated based on their year of birth, rounded down to the nearest whole year.
        - The current date and time are used to calculate the person's age.
        """
        dob = datetime.datetime.strptime(self.dob, '%Y-%m-%d').date()
        today = datetime.date.today()
        age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
        return age
```

```python
from flask import Flask, request
import json
import datetime

app = Flask(__name__)

class Person:
    def __init__(self, name, dob):
        self.name = name
        self.dob = dob
    
    def get_approx_age(self):
        """
        Returns the approximate age of the person based on their date of birth.
        
        Assumptions:
        - The person's date of birth is provided as a string in the format 'YYYY-MM-DD'.
        - The person's age is calculated based on their year of birth, rounded down to the nearest whole year.
        - The current date and time are used to calculate the person's age.
        """
        dob = datetime.datetime.strptime(self.dob, '%Y-%m-%d').date()
        today = datetime.date.today()
        age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
        return age

@app.route('/age', methods=['GET'])
def get_age():
    """
    Returns the approximate age of a person based on their name and date of birth.
    
    Parameters:
    - name: The name of the person as a string.
    - dob: The person's date of birth as a string in the format 'YYYY-MM-DD'.
    
    Returns:
    - The approximate age of the person as a JSON object, with the following format:
        {"name": <name>, "age": <age>}
    """
    name = request.args.get('name')
    dob = request.args.get('dob')
    
    if name is None or dob is None:
        return json.dumps({"error": "Name and date of birth must be provided."}), 400
    
    person = Person(name, dob)
    age = person.get_approx_age()
    
    return json.dumps({"name": name, "age": age}), 200

if __name__ == '__main__':
    app.run(debug=True)
```
